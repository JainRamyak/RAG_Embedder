# RAG Embedder — Interactive Demo
Explore local embeddings without spending a single API dollar.

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from embedder import get_embedder

embedder = get_embedder()
print(f'Embedder ready — dimension: {embedder.dimension}')
```

## 1. Single Text → Vector

```python
vec = embedder.embed_text('What is machine learning?')
print(f'Vector length : {len(vec)}')
print(f'First 10 dims : {[round(v, 4) for v in vec[:10]]}')
```

## 2. Similarity Heatmap

```python
sentences = [
    'The cat sat on the mat',
    'A kitten rested on the rug',
    'Stock markets fell today',
    'Investors saw major losses',
    'Deep learning uses neural networks',
    'AI models learn from data',
]

vectors = embedder.embed_batch(sentences)
vecs = np.array(vectors)

# cosine similarity matrix
norms = np.linalg.norm(vecs, axis=1, keepdims=True)
normed = vecs / norms
sim_matrix = normed @ normed.T

short_labels = [
    'Cat/mat', 'Kitten/rug',
    'Markets fell', 'Investors',
    'Deep learning', 'AI models'
]

plt.figure(figsize=(8, 6))
sns.heatmap(
    sim_matrix,
    annot=True, fmt='.2f',
    xticklabels=short_labels,
    yticklabels=short_labels,
    cmap='RdYlGn', vmin=0, vmax=1
)
plt.title('Cosine Similarity Heatmap', fontsize=14, pad=12)
plt.tight_layout()
plt.savefig('similarity_heatmap.png', dpi=150)
plt.show()
print('Saved → notebooks/similarity_heatmap.png')
```

## 3. Try Your Own Pairs

```python
pairs = [
    ('I love pizza', 'Pizza is my favourite food'),
    ('I love pizza', 'The economy is struggling'),
    ('Python is great for data science', 'NumPy and pandas are powerful'),
]

print(f'{"Text A":<40} {"Text B":<40} Score')
print('-' * 90)
for a, b in pairs:
    score = embedder.similarity(a, b)
    print(f'{a:<40} {b:<40} {score:.4f}')
```
